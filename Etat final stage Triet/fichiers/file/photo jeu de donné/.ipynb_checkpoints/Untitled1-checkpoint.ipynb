```python
import pandas as pd
import psycopg2
import os
from io import StringIO
```

```python
photo = pd.read_excel('valeurs mesurées2.xlsx', sheet_name = 0, header = 3)
data_sensor = pd.read_excel('valeurs mesurées2.xlsx', sheet_name = 1)
```

```python
def connect() :
    argv = {"host" : "localhost",
            "dbname" : "lmsm",
            "user" : "iventre",
            "password" : "lmsm"}
    connect = psycopg2.connect(**argv)
    print('connect success')
    return connect
conn = connect()
cur = conn.cursor()
```

```python
try :
    cols = list([x.split('.')[0] for x in data_sensor.columns])
    param = list(data_sensor.iloc[0, :])

    lst = [[col, par] for col, par in list(zip(cols, param))]
    val = lst[1 : -1 : 2]
    k = 0
    for i in val :
        data = data_sensor[data_sensor.columns[k:k + 2]][1:].reset_index().drop('index', axis = 1)
        lt = pd.DataFrame({'Position' : [i[0]] * len(data), 'Parameter' : [i[1]] * len(data)})
        result = pd.concat([lt, data], axis = 1)
        buffer = StringIO()
        result.to_csv(buffer, index=False, header=False)
        buffer.seek(0)
        cur.copy_from(buffer, 'a', sep=",", null = "") 
        conn.commit() 
        print(f"copy_from_stringio() {i[0]} - {i[1]} done")
        k += 2
except (Exception, psycopg2.DatabaseError) as error:
    print("Error: %s" % error)
    conn.rollback()
    cursor.close()
```
